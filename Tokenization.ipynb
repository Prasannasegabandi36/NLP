```python
corpus="""Hello this is the Complete NLP,ML,DL course.

please do watch the entire course! Subscribe Krish naik channel.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## senetnce ---> Paragraphs
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)
```

```python
documents
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Tokenization
# Paragraph --> words
## sentence --> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
## for puncuations it ll use
wordpunct_tokenize(corpus)
```

```python
# full stop never treat as seperate word only for last word full stop will come with seperate
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
