```python
from nltk.stem import   WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
'''
POS -Noun-n
verb-v
adj-a
adverb-r
'''
lemmatizer.lemmatize("running",pos='v')
```

```python
words = ["playing", "played", "player", "studies", "studying", "connection"]
```

```python
for word in words:
    print(word+'--->'+lemmatizer.lemmatize(word,post='n'))
```
