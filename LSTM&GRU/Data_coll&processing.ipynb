```python
import nltk
nltk.download('gutenberg')
from nltk.corpus import gutenberg
import pandas as pd
## load the dataset
data=gutenberg.raw('shakespeare-hamlet.text')
## save to file 
with open('hamlet.txt','w') as file:
    file.write(data)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split\

#load the datset
with open('hamlet.txt','r')as file:
    text=file.read().lower()
# tokeinze text
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
total_words=len(tokenizer.word_index)+1
total_words
```

```python
tokenizer.word_index
```

```python
## create input sequences
input_sequences=[]
for line in text.split('\n'):
    token_list=tokenizer.text_to_sequences([line])[0]
    for i in range(1,len(token_list)):
        n_gram_sequence=token_list[:i+1]
        input_sequences.append(n_gram_sequence)
```

```python
input_sequences
```

```python
max_sequence_len=max([len(x) for x in input_sequences])
max_sequence_len
```

```python
input_sequences-np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))
input_sequences
```

```python
##create predicitors and lebel
import tensorflow as tf
x,y=input_sequences[:,:-1],input_sequences[:,-1]
```

```python
x
```

```python
y
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=total_words)
y
```

```python
#split the data into traning and testing sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#define early stopping
from tensorflow.keras.callbacks import EarlyStopping
early_stopping =EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
```

```python
#Train our LSTM RNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout

#define the model
model=Sequential()
model.add(Embedding(total_words,100,input_length=max_sequence_len))
model.add(LSTM(150, return_sequences=True))
model.add(Dropout=0.2)
model.add(LSTM(100))
model.add(Dense(total_words,activation="softmax"))
##complie the model
model.complete(loss="categorical_crossentropy",optimizer="adam",metrices=['accuracy'])
model.summary()
```

```python
## train the model
history=model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test),verbose=1)
```

```python
# Predict from lstm
def predict_next_word(model,tokenizer,test,max_sequence_len):
    token_list=tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_sequence_len:
        token_list = token_list[-(max_sequence_len):]
    token_list = pad_sequences([token_list],maxlen=max_sequence_len-1,padding='pre')
    predicted=model.predict(token_list,verbose=0)
    predicted_word_index=np/argmax(predicted,axis=1)
    for word,index in tokenizer.word_index.items():
        if index == predicted_word_index:
            return word
    return None
```
